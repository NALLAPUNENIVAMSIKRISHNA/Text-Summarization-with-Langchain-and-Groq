# 2 nd technique - Prompt template text summarization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate
generictemplate="""Write the summary of the following speech.
Speech:{speech}
Translate the precise summary to {language}
"""

prompt=PromptTemplate(
    input_variables=["speech", "language"],
    template=generictemplate
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="Hindi")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'Hindi'})
summary
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'Telugu'})
summary
```

In the text we are sending from pdf's to prompt template we are having challenges like doc becomes very big now we use another technique

The larger documents will be divided into smaller chunks -> prompt message 1-> llm ->summaries ( summary 1, summary 2 , summary 3 ,...) -> prompt template2 -> llm -> final summary

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load_and_split()
docs
```

```python
template="""
Write a concise and short summary of the following speech,
Speech :{text}
"""
prompt=PromptTemplate(input_variables=["text"],template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
output_summary=chain.run(docs)
output_summary
```

# Refine Chain summarization

We are having doc's and then we divide them into chunks -> prompt -> llm -> summary                                                      
chunk 1->prompt->llm->summary this summary will be given to chunk2                                                                         
chunk 2-> prompt ->llm -> summary this summary given to chunk 3                                                                            
chunk 3-> prompt ->llm -> summary .....  -> final summary

The refine documents chain constructs a response by looping over the input documents and iteratively updating its answer. For each document, it passes all non-document inputs, the current document, and the latest intermediate answer to an LLM chain to get a new answer.

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)
output_summary=chain.run(final_documents)
output_summary
```
